```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load your CSV
df = pd.read_csv("Restaurant.csv")
print(df.head())   # Peek at data

# Bar Plot
sns.barplot(x="Product", y="Quantity", data=df)
plt.title("Average Sales by Product")
plt.show()

# Line Plot
sns.lineplot(x="Product", y="Quantity", data=df, marker="o")
plt.title("Sales Over Product")
plt.xticks(rotation=45)
plt.show()

#Scatter Plot
sns.scatterplot(x="Product", y="Price", data=df)
plt.title("Products vs Price")
plt.show()

#Correlation Heatmap

corr = df.corr(numeric_only=True)  # only numeric columns
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
