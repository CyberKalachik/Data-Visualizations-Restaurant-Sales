```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Restaurant.csv")

# Quick look
print(df.head())
print(df.columns)

#Bar Chart

df.groupby("Product")["Quantity"].sum().plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()

#Bar Chart
df.groupby("City")["Quantity"].sum().plot(kind="bar")
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Quantity")
plt.show()

#Histogram
df["Quantity"].plot(kind="hist", bins=20, edgecolor="black")
plt.title("Sales Distribution")
plt.xlabel("Quantity")
plt.show()
```
